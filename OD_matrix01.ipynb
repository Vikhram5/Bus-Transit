```python
# Define the bus stages in the desired order
bus_stages = [
    'T.NAGAR', 'SAIDAPET', 'ANNA UNIVE', 'WPTC', 'SRP TOOLS',
    'KANDANCHAV', 'THORAIPAKK', 'M K CHAVADI', 'KARAPAKKAM',
    'SHOLINGANALLUR', 'KUMARAN NAGAR', 'CHEMMANCHERY', 'NAVALUR',
    'SIPCIOT', 'CHURCH', 'PAL CHEM', 'HINDUSTAN', 'KELAMBAKKAM',
    'KOMAN NAGAR', 'ENGG', 'THENGAMMAL', 'THIRUPPORU'
]

# Create an empty 2D matrix filled with zeros
num_stages = len(bus_stages)
od_matrix = [[0 for _ in range(num_stages)] for _ in range(num_stages)]

# Populate the matrix with adult counts from your dataset
# Assuming you have a DataFrame named df with your data
for _, row in df.iterrows():
    from_stage = row['From Stage']
    to_stage = row['To Stage']
    adult_count = row['Adult']
    from_index = bus_stages.index(from_stage)
    to_index = bus_stages.index(to_stage)
    od_matrix[from_index][to_index] += adult_count

# Display the OD matrix
print("OD Matrix:")
for row in od_matrix:
    print(row)
```

```python
# Define the bus stops
bus_stops = [
    'Stop A', 'Stop B', 'Stop C', 'Stop D', 'Stop E'
]

# Define the number of passengers traveling between each pair of stops
passenger_counts = [
    [0, 10, 5, 0, 20],  # From Stop A
    [10, 0, 15, 25, 0],  # From Stop B
    [5, 15, 0, 0, 30],  # From Stop C
    [0, 25, 0, 0, 10],  # From Stop D
    [20, 0, 30, 10, 0]  # From Stop E
]

# Create an empty OD matrix
od_matrix = [[0 for _ in range(len(bus_stops))] for _ in range(len(bus_stops))]

# Populate the matrix with passenger counts
for i in range(len(bus_stops)):
    for j in range(len(bus_stops)):
        od_matrix[i][j] = passenger_counts[i][j]

# Display the OD matrix
print("OD Matrix:")
for row in od_matrix:
    print(row)
```

```python
import pandas as pd

# Convert the OD matrix to a pandas DataFrame
od_df = pd.DataFrame(od_matrix, index=bus_stops, columns=bus_stops)

# Print the DataFrame
print("OD Matrix:")
print(od_df)

```

```python
import pandas as pd

# Define the bus stages in the desired order
bus_stage = [
    'T.NAGAR', 'SAIDAPET', 'ANNA UNIVE', 'WPTC', 'SRP TOOLS',
    'KANDANCHAV', 'THORAIPAKK', 'M K CHAVADI', 'KARAPAKKAM',
    'SHOLINGANALLUR', 'KUMARAN NAGAR', 'CHEMMANCHERY', 'NAVALUR',
    'SIPCOT', 'CHURCH', 'PAL CHEM', 'HINDUSTAN', 'KELAMBAKKAM',
    'KOMAN NAGAR', 'ENGG', 'THENGAMMAL', 'THIRUPPORU'
]

# Assuming you have a DataFrame named df with your data
# Convert 'From Stage' and 'To Stage' columns to categorical with specified order
df['From Stage'] = pd.Categorical(df['From Stage'], categories=bus_stage, ordered=True)
df['To Stage'] = pd.Categorical(df['To Stage'], categories=bus_stage, ordered=True)

# Create the OD matrix
od_matrix = df.groupby(['From Stage', 'To Stage'])['Adult'].sum().unstack(fill_value=0)

# Display the OD matrix
print("OD Matrix:")
od_matrix['Boarding'] = od_matrix.sum(axis=1)
od_matrix.loc['ALIGHTING'] = od_matrix.sum(axis=0)
od_matrix
od_matrix.info()
```
